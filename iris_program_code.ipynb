```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n=['sepal-length','sepal-width','petal-length','petal width','class']
dataset=pd.read_csv(url,names=n)
```

```python
x=dataset.iloc[:,0:4]
y=dataset.iloc[:,4]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
ts_score=[]
for j in range(100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=j,test_size=.15)
    lr=KNeighborsClassifier().fit(x_train,y_train)
    ts_score.append(lr.score(x_test,y_test))
J=ts_score.index(np.max(ts_score))
```

```python
J
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=.15)
```

```python
x_train
```

```python
knn=KNeighborsClassifier(n_neighbors=45)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
#to find the k value
error=[]
#it is starting from 3 becoz k cant be 1, k cant be 2 even so 3 and we dont need even no so step is 2
for i in range(3,100,2):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_y_test=knn.predict(x_test)
    error.append(np.mean(pred_y_test!=y_test))
```

```python
error
```

```python
plt.plot(range(3,100,2),error,color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.title("error rate k value")
plt.xlabel('k value')
plt.ylabel('mean error')
```
